```python
#pivot the network_distance dataframe.
#Here the pivot table is transformed to numpy 2D array such as spopt.locate expected.
ntw_time_piv = network_travel_time.pivot_table(values="DriveTime", index="DemandID", columns="StationID")
ntw_time_piv
```

```python
#setting the demand weight
#ai is the demand weight/call frequency at a location
ai = demand_points_gdf['CallFreq'].to_numpy()

service_time = 20 #minutes
p_facility = 15
```
